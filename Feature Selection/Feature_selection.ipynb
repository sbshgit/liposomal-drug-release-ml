Heatmap of the absolute Spearmanâ€™s Rank correlation between  the initial 24 input features

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
datafile = pd.read_excel("24_input_features.xlsx")
X = datafile.drop(columns=['Formulation_Number','Drug_Lipid_Group','Release','P_Type_HSPC', 'P_Type_lecithin', 'P_Type_DMPC', 'P_Type_DPPC', 'Prep_Method-Thin-fim-evaporation', 'Prep_Method-Microfluidics', 'Prep_Method-Reverse-phase-evaporation'])
X = X.loc[:, (X != X.iloc[0]).any()]
correlation = X.corr(method='spearman')
sns.clustermap(round(np.abs(correlation),2),vmin=0, vmax=1,  annot=True, figsize=(25,25),  cbar=True, cmap='PiYG'
               , cbar_pos=(0, .2, .03, .4),  fmt=".2f",   linecolor='black', linewidths=0.5)
plt.savefig('ArticleFigures/clustermap1', dpi=600, format = 'png', transparent=True, bbox_inches='tight')
plt.show()
```

```python
def run_nested_cross_validation(name, CV):

  if __name__ == '__main__':
    model_instance =  nested_cross_validation(model_type = name)
    model_instance.input_target()
    model_instance.cross_validation(CV)
    model_instance.results()
    model_instance.best_model()
    model_instance.CV_dataset.to_pickle("Feature_selection_NGB_Results/20_feat_NGB.pkl", compression='infer', protocol=5, storage_options=None) # save dataframe as pickle file
    with open('Feature_selection_NGB_Model_Pickel_Files/20_feat_NGB_model.pkl', 'wb') as file: # Save the Model to pickle file
          pickle.dump(model_instance.best_model, file)
    model_instance.CV_dataset.to_excel("Feature_selection_NGB_Results/20_feat_NGB.xlsx", index=False)  # Save the DataFrame as an Excel file
    
    return model_instance.CV_dataset, model_instance.best_model
```
