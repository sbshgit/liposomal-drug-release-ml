```python
def run_nested_cross_validation(name, CV):

  if __name__ == '__main__':
    model_instance =  nested_cross_validation(model_type = name)
    model_instance.input_target()
    model_instance.cross_validation(CV)
    model_instance.results()
    model_instance.best_model()
    model_instance.CV_dataset.to_pickle("nested_cross_validation_results/24_feat_"+str(name)+".pkl", compression='infer', protocol=5, storage_options=None) # save dataframe as pickle file
    with open('Trained_models_pickel_files/24_feat_'+str(name)+'_model.pkl', 'wb') as file: # Save the Model to pickle file
          pickle.dump(model_instance.best_model, file)
    model_instance.CV_dataset.to_excel("nested_cross_validation_results/24_feat_" + str(name) + ".xlsx", index=False)  # Save the DataFrame as an Excel file
    
    return model_instance.CV_dataset, model_instance.best_model
```

Box Plot of The Models Results

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to extract and compile a results dataframe as well as an absolute error array for all models in NESTED_CV_results pickle files
def extraction_all(name):
    file_path = os.path.join( 'nested_cross_validation_results',f'24_feat_{name}.pkl')
    df = pd.read_pickle(file_path, compression='infer', storage_options=None)
    
    list_of_dataframes = []

    for n in range(10):
        dataframe = pd.DataFrame(df['Time'][n], columns=['Time'])
        dataframe['Experimental_Release'] = df['Experimental_Release'][n]
        dataframe['Predicted_Release'] = df['Predicted_Release'][n]
        dataframe['Formulation_Number'] = df['Formulation_Number'][n]
        dataframe['Absolute_Error'] = abs(dataframe['Experimental_Release'] - dataframe['Predicted_Release'])
        list_of_dataframes.append(dataframe)

    dataframe_all = pd.concat(list_of_dataframes, axis=0, ignore_index=True)
    return dataframe_all

# Extracting data for all models
model_names = ['lasso', 'KNN', 'PLS', 'SVR', 'DT', 'RF', 'LGBM', 'XGB', 'NGB']
all_models = {name: extraction_all(name) for name in model_names}

# Combine absolute errors into a single DataFrame
all_ae = pd.DataFrame({name: model['Absolute_Error'] for name, model in all_models.items()})

# Sort columns by mean absolute error
sorted_index = all_ae.mean().sort_values().index
df_sorted = all_ae[sorted_index]

# Save the sorted DataFrame to an Excel file
df_sorted.to_excel("Box_plot_dataset.xlsx")

# Describe the DataFrame
print(df_sorted.describe())
```

```python
# Plotting the box plot
plt.figure(figsize=(12, 6))
palette = sns.color_palette("pastel")
sns.set_style("white")

boxplot = sns.boxplot(data=df_sorted, palette=palette, saturation=0.8,
                      boxprops=dict(linewidth=1.0, edgecolor='black', alpha=0.8),
                      whiskerprops=dict(linewidth=1.0, color='black'),
                      capprops=dict(linewidth=1.0, color='black'),
                      flierprops=dict(marker="d", markerfacecolor="black", markeredgecolor="black",
                                      markersize=0.5, alpha=0.2),
                      medianprops=dict(color="black", linewidth=1.0, linestyle='--'),
                      showmeans=True,
                      meanprops=dict(marker="o", markerfacecolor="black", alpha=1.0,
                                     markeredgecolor="black", markersize=4, linewidth=0.05, zorder=10))

# Add each datapoint
sns.stripplot(data=df_sorted, marker="o", edgecolor='white', alpha=0.3, size=1.5, linewidth=0.3, color='black', zorder=0)

# Titles and labels
boxplot.set_xlabel("Model index", fontsize=16, color='black', weight="bold")
boxplot.set_ylabel("Absolute error (AE)", fontsize=16, color='black', weight="bold")

# y-axis limits and interval
boxplot.set(ylim=(-0.02, 2.02), yticks=np.arange(0,85,5))

# x-axis rotation and text color
boxplot.set_xticklabels(boxplot.get_xticklabels(), rotation=0, color='black', fontsize=12, weight="bold")

# x-axis and y-axis tick color
boxplot.tick_params(colors='black', which='both')  # 'both' refers to minor and major axes
boxplot.tick_params(bottom=False, left=True)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, weight="bold")
plt.tight_layout()
plt.savefig('ArticleFigures/BoxPLot', format='png', transparent=True)
plt.show()
```
