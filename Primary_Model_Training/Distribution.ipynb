Distribution of the fractional drug release values for the 24-feature liposome datasets.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel("24_input_features.xlsx")
f, boxplot = plt.subplots(figsize=(10,6))
ax = sns.histplot(df['Release'], binwidth=1, alpha=0.6 , color='mediumvioletred', kde=True)
ax.lines[0].set_color('black')
plt.ylabel("Count", fontsize=14, color='black',  weight="bold")
plt.xlabel("Release", fontsize=14, color='black', weight="bold")
plt.xticks(np.arange(0, 90,5)) 
plt.yticks(np.arange(0,26))
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.tight_layout()
plt.savefig('ArticleFigures/HistPlot1', dpi=600, format = 'png', transparent=True, bbox_inches='tight')
plt.show()
```

Distribution of drug-phospholipid groups in the 24-feature dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel("24_input_features.xlsx")
f, boxplot = plt.subplots(figsize=(10,6))
ax = sns.histplot(df["Drug_Lipid_Group"], binwidth=1, alpha=0.6 , color='mediumvioletred', kde=True)
ax.lines[0].set_color('black')
plt.ylabel("Count", fontsize=14, color='black',  weight="bold")
plt.yticks(np.arange(0,300,20))
plt.yticks(fontsize=10)
plt.xticks(rotation=90, fontsize=10)
plt.tight_layout()
plt.savefig('ArticleFigures/HistPlot2', dpi=600, format = 'png', transparent=True, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'DPPC', 'HSPC', 'Lecithin', 'DMPC'
sizes = [130, 67, 53, 10]
colors = "tomato", "coral", "lightsalmon", "peachpuff"
f, boxplot = plt.subplots(figsize=(10,6))
plt.pie(x=sizes, labels=labels, colors=colors, autopct='%1.1f%%', radius=1.15,
         textprops={'fontsize':9, 'color':'black', 'weight':'bold'} )
plt.title('Phospholipid Type', fontsize=12, color='black',  weight="bold", x=0.5, y=1.02)
plt.legend( bbox_to_anchor=(-0.01, 1))
plt.tight_layout()
plt.savefig('ArticleFigures/Pie1', dpi=600, format = 'png', transparent=True, bbox_inches='tight')
plt.show()
```
